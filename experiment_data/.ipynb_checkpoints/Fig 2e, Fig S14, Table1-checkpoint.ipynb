```python
def transf_pos(x,K=2.0,y0=0.001,ym=1.5,n=2):
    result=ym*((x/K)**n)/(1+(x/K)**n)+ym*y0
    return result

def transf_neg(x,K=2.0,y0=0.001,ym=1.5,n=2):
    result=ym/(1+(x/K)**n)+y0*ym
    return result
```

```python
def channel_median(facsdata_list,channel,flip_tag=True):
    channel_median=[]
    
    if flip_tag==True:
        for i in np.flip(np.arange(len(facsdata_list))):
            channel_median.append(np.median(facsdata_list[i][:,channel]))
    
    else:
        for i in np.arange(0,len(facsdata_list)):
            channel_median.append(np.median(facsdata_list[i][:,channel]))
        
    return np.array(channel_median)
```

```python
# the following section is for sensitivity analysis as in Fig S14

from SALib.sample import saltelli
from SALib.analyze import sobol
```

```python
def fiterr_(param,y0):
    y_param_=transf_pos(AHL_levels,*param)
    y_err=np.sqrt(np.sum((y_param_-y0)**2))
    
    return y_err
```

```python
ydata=mean_m7 # can also be mean_m8, mean_m7, mean_m40
problem={
    'num_vars':4,
    'names':['K','y0','ym','n'],
    'bounds':[[0,100],  # 100uM is the maximum OC6 level used in experiment
              [0,ydata[0]/ydata[-1]],
              [ydata[-1],1e5],
              [0.2,1.0]]
    }
param_values = saltelli.sample(problem, 2000)
Y=np.zeros([param_values.shape[0]])

for i in range(0,len(param_values)):
    Y[i]=fiterr_(param_values[i],ydata)

# correspondingly, Si_m7, Si_m8, Si_m40
Si_m7=sobol.analyze(problem,Y)
print(Si_m7['S1'])
```

```python
# FigS14
fig=plt.figure(figsize=(5,3.5))
ax=fig.add_subplot(111)

ax.bar(x=np.linspace(0,3,4),height=Si_m7['S1'],width=0.18,color='cornflowerblue',alpha=0.8,label='mut7')
ax.bar(x=np.linspace(0,3,4)+0.19,height=Si_m8['S1'],width=0.18,color='coral',alpha=0.8,label='mut8')
ax.bar(x=np.linspace(0,3,4)+0.19*2,height=Si_m15['S1'],width=0.18,color='grey',alpha=0.8,label='mut15')
ax.bar(x=np.linspace(0,3,4)+0.19*3,height=Si_m40['S1'],width=0.18,color='firebrick',alpha=0.8,label='mut40')
ax.legend(loc='upper left',fontsize=12)

ax.tick_params(direction='in',labelsize=11)

ax.set_xticks([0.3,1.3,2.3,3.3])
ax.set_xticklabels(['$K_d$','$\\beta_0$','$\\beta_m$','n'],size=12.5)

ax.set_ylabel('first order sensitivity',fontsize=12)
ax.set_xlabel('fitting parameters',fontsize=12)
```
